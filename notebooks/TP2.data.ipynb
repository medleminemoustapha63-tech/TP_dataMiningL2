Detection par IQR

# SIMULATION DE PRIX (EN Millions MRU)

```python
prix= pd.Series([1.2 , 1.5 , 1.1 , 120.0 , 1.4 , 1.3 , 0.9 , 85.0 , 1.2])
Q1=prix.quantile(0.25)
Q3=prix.quantile(0.75)
IQR=Q3-Q1
# definition de bornes
borne_inf = Q1 - 1.5 * IQR
borne_sup = Q3 + 1.5 * IQR
outliers = prix [( prix < borne_inf ) | ( prix > borne_sup ) ]
print (f" Valeurs ␣ aberrantes ␣ detectees ␣:␣{ outliers . values }")
```

CALCULER Z-SCORE

X=120,
μ=23,73 #MOYENNE,
σ=43,
Z=(x −μ)/σ),
DONC Z=2,23 INTERPRETATION:120 TRES PETIT OUTLIER


```python
prix=prix.clip(lower=borne_inf,upper=borne_sup)
```
